# ¿Qué es un "Modern Data Stack"?
Un "modern data stack" (conjunto de datos moderno) es un enfoque contemporáneo para gestionar y analizar datos utilizando una combinación de herramientas y tecnologías modernas, generalmente basadas en la nube. Este stack se caracteriza por ser escalable, flexible y modular, lo que permite a las organizaciones manejar grandes volúmenes de datos de manera eficiente y obtener insights accionables rápidamente.



## 1. Fuentes de Datos (Sources)

Las fuentes de datos son los orígenes desde los cuales se extraen los datos. Pueden ser internas o externas y presentan datos en diversos formatos y estructuras.

Tipos de Fuentes:

- Bases de Datos Relacionales: Como PostgreSQL, MySQL.

- APIs: Servicios web que proporcionan datos en tiempo real.

- Archivos: CSV, JSON, XML, hojas de cálculo.

- Sistemas de Streaming: Kafka, Kinesis.

- Dispositivos IoT y Sensores: Generan datos en tiempo real.

- Aplicaciones Empresariales: CRM, ERP, sistemas de gestión.

## 2. Bases de Datos OLTP (Online Transaction Processing)

 * Son sistemas de bases de datos diseñados para manejar grandes cantidades de transacciones en tiempo real.
  
  * Resuelven el problema de gestionar múltiples transacciones simultáneas sin conflictos y con seguridad. asegurando que las operaciones cotidianas, como inserciones, actualizaciones y eliminaciones de datos, se procesen de manera rápida y eficiente, garantizando la integridad de los datos. 
  

  **¿Por qué es útil para un data scientist saber de Bases OLTP?**

  * Aunque un data scientist no siempre está involucrado en el diseño o mantenimiento de bases OLTP, muchas veces sus análisis comienzan con datos almacenados en estas bases transaccionales. 

  *  Entender cómo se estructuran y operan las bases OLTP permite al data scientist extraer datos más eficientemente, escribir consultas optimizadas y comprender la dinámica de los datos en tiempo real.

## 2. Ingesta de Datos (Ingestion)

* Es el proceso de recopilar, integrar y cargar datos de diversas fuentes (APIs, archivos, bases de datos OLTP, entre otras) en un sistema centralizado, como un data lake o un data warehouse. 

**¿Por qué es útil para un data scientist?**: 

* La ingesta de datos asegura que la información de diversas fuentes (APIs, archivos, bases de datos) llegue a donde debe estar. 

* Logra eficiencia al permitir la carga de datos incrementales de forma sencilla-


##  3. Data Warehouse (OLAP)

* Son bases de datos diseñadas para la consulta y análisis de grandes volúmenes de datos históricos. 

* A diferencia de las bases de datos OLTP, que priorizan la escritura rápida, los data warehouses OLAP están optimizados para la lectura y consulta eficiente de datos, permitiendo ejecutar consultas complejas y agregaciones sobre grandes conjuntos de datos. 

* Estas bases son fundamentales para generar informes, dashboards y realizar análisis avanzados que ayuden a la toma de decisiones estratégicas.

**¿Por qué es útil saber de OLAP para un data scientist?**: 

* Porque permiten procesar grandes volúmenes de datos en un entorno optimizado para consultas analíticas complejas.



## 4. Orquestación (Orchestration)

La orquestación coordina y automatiza los diversos procesos y tareas en el pipeline de datos, asegurando que ocurran en el orden correcto y manejando dependencias.

**¿Por qué es útil para un data scientist?**

* Para garantizar que los datos y los modelos se actualicen sin intervención manual.

* Para facilitar la gestión de pipelines complejos y asegurar que los análisis se realicen con datos actualizados
